```python
# 1. Import libraries
import pandas as pd
import matplotlib.pyplot as plt

# 2. Load dataset
df = pd.read_csv('/content/mock_kaggle.csv')   # This path works in Colab

#renamed language
df.rename(columns={'data': 'date', 'venda': 'sales', 'estoque': 'stock', 'preco': 'price'}, inplace=True)

# 3. View first 5 rows
print(df.head())

# 4. View basic info
print(df.info())

# 5. Check missing values
print(df.isnull().sum())

# 1. Check data types
print(df.dtypes)

# Convert 'date' column from object to datetime
df['date'] = pd.to_datetime(df['date'])

# Check again
print(df.dtypes)
# 3. Plot Sales Over Time

import matplotlib.pyplot as plt

# Set figure size
plt.figure(figsize=(12,6))

# Plot
plt.plot(df['date'], df['sales'], marker='o')

# Add title and labels
plt.title('Sales Trend Over Time', fontsize=16)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Sales', fontsize=14)
plt.grid(True)

# Show plot
plt.show()
# Summary statistics
print(df.describe())
# Check for duplicate rows
print("Total duplicate rows:", df.duplicated().sum())
# Daily sales histogram
plt.figure(figsize=(8,5))
plt.hist(df['sales'], bins=20, color='skyblue', edgecolor='black')
plt.title('Sales Distribution')
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
import numpy as np
from sklearn.linear_model import LinearRegression

# Sort data by date just to be sure
df = df.sort_values(by='date')

# Reset index
df = df.reset_index(drop=True)

# Create "day number" column (0, 1, 2, ...)
df['day_number'] = np.arange(len(df))

# Check it
print(df[['date', 'day_number', 'sales']].head())
from sklearn.model_selection import train_test_split

# Define X and y
X = df[['day_number']]
y = df['sales']

# Split into train and test sets (80/20)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

# Print shapes
print("Train size:", X_train.shape[0])
print("Test size:", X_test.shape[0])
# Initialize the model
model = LinearRegression()

# Fit the model
model.fit(X_train, y_train)

# Predict on training and test data
train_preds = model.predict(X_train)
test_preds = model.predict(X_test)
# Plot actual vs predicted sales
plt.figure(figsize=(12,6))

# Plot actual test sales
plt.plot(X_test, y_test.values, label='Actual Sales', marker='o')

# Plot predicted test sales
plt.plot(X_test, test_preds, label='Predicted Sales', marker='x')

# Add labels
plt.title('Actual vs Predicted Sales (Test Data)', fontsize=16)
plt.xlabel('Day Number', fontsize=14)
plt.ylabel('Sales', fontsize=14)
plt.legend()
plt.grid(True)
plt.show()
# Add rolling average (past 7 days)
df['rolling_avg'] = df['sales'].rolling(window=7).mean()
# Define the threshold for sales drop (e.g., 30%)
drop_threshold = 0.30

# Create a column to track sales drop alerts
df['sales_drop_alert'] = (df['sales'] < df['rolling_avg'] * (1 - drop_threshold))

# Display rows where alert is triggered
alerts = df[df['sales_drop_alert']]

# Show alerts
print(alerts[['date', 'sales', 'rolling_avg', 'sales_drop_alert']])
# Show all triggered alerts
print(df[df['sales_drop_alert']][['date', 'sales', 'rolling_avg', 'sales_drop_alert']])
def generate_alerts(df, threshold=0.30):
    df['rolling_avg'] = df['sales'].rolling(window=7).mean()
    df['sales_drop_alert'] = df['sales'] < df['rolling_avg'] * (1 - threshold)
    return df[df['sales_drop_alert']][['date', 'sales', 'rolling_avg', 'sales_drop_alert']]
# Try with a 20% threshold
alerts_20 = generate_alerts(df.copy(), threshold=0.20)

# Try with a 40% threshold
alerts_40 = generate_alerts(df.copy(), threshold=0.40)
def label_severity(row):
    drop_ratio = 1 - (row['sales'] / row['rolling_avg'])
    if drop_ratio >= 0.5:
        return 'Severe'
    elif drop_ratio >= 0.3:
        return 'Moderate'
    elif drop_ratio >= 0.15:
        return 'Mild'
    else:
        return None

# Apply after rolling average
df['severity'] = df.apply(lambda row: label_severity(row) if row['rolling_avg'] != 0 else None, axis=1)
alert_severity = df[df['severity'].notna()][['date', 'sales', 'rolling_avg', 'severity']]
print(alert_severity)
import matplotlib.pyplot as plt

# Plot actual sales and rolling average
plt.figure(figsize=(14, 6))
plt.plot(df['date'], df['sales'], label='Actual Sales', color='blue')
plt.plot(df['date'], df['rolling_avg'], label='7-Day Rolling Avg', color='orange')

# Highlight points with alerts
alert_points = df[df['severity'].notna()]
plt.scatter(alert_points['date'], alert_points['sales'], color='red', label='Sales Drop Alert', zorder=5)

plt.xlabel('Date')
plt.ylabel('Sales')
plt.title('Sales Trends with Drop Alerts')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
alerts_df = generate_alerts(df.copy(), threshold=0.30)
print(alerts_df.sort_values('date'))


import pandas as pd

# Set Pandas to display all rows
pd.set_option('display.max_rows', None)

# Print the DataFrame (this will show all rows now)
print(df)




```
